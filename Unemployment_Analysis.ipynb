```python
#importing liraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
#loading dataset
df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
df
```

```python
#briefing dataset
df.describe()
```

```python
#Data Standardization
df.isnull().sum()
```

```python
#Renaming the columns
df.columns = ["State","Date","Frequency","Estimated Unemployment Rate","Estimated Employed",
              "Estimated Labour Participation Rate","Region","Longitude","Latitude"]
```

```python
#checking
df.head()
```

```python
#Correlation coefficient
sns.heatmap(df.corr(),annot=True)
```

```python
#Data visualization of Estimated Employees according to region
plt.title("Unemployment in India")
sns.histplot(x="Estimated Employed",hue="Region",data=df)
plt.show()
```

```python
#Visualizing main parameter and its feature across the dataset
plt.title("Unemployment in India")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=df)
plt.show()
```

```python
#Data Visualization of main feature
unemployment = df[["State","Region","Estimated Unemployment Rate"]]
fig = px.sunburst(unemployment,path=["Region","State"],values="Estimated Unemployment Rate",color_continuous_scale="RdY1Gn")
fig.show()
```

```python
#Statewise Unemployement Rate to better understand the then scenario across India
plt.title("Unemployment in India")
sns.barplot(y="Estimated Unemployment Rate",x="State",data=df)
plt.xticks(rotation = 90)
plt.show()
```
